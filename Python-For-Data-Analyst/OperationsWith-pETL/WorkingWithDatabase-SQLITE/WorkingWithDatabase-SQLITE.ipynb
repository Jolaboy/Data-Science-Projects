STEPS TO PERFORM IN SQLITE
- navigate to the datasets directory
- create a database :
sqlite3 index.db
- attach a database:
ATTACH DATABASE 'index.db' As 'index';
- create a table:
CREATE TABLE NYSE (
	Date TEXT,
   	Open REAL NOT NULL,
	High REAL NOT NULL,
	Low  REAL NOT NULL,
    Close REAL NOT NULL,
    AdjClose REAL NOT NULL,
    Volume REAL NOT NULL
   );

- check if table is there:
.tables

- insert csv data to the table:
.mode csv
.import NYA.csv NYSE

.mode column
.headers on
select * from NYSE;

- drop database :DETACH DATABASE dbname

```python
connection = sqlite3.connect('datasets/index.db')
```

```python
index_data = etl.fromdb(connection, 'SELECT * FROM NYSE')
```

```python
index_data.display()
```

```python
etl.nrows(index_data)
```

#### BASIC TRANSFORMATIONS

```python
sliced_data = etl.rowslice(index_data, 10)
```

```python
sliced_data.displayall()
```

```python
feb_data_2018 = etl.rowslice(index_data, 21, 40)
```

```python
feb_data_2018.displayall()
```

```python
alternate_data_feb_2018 = etl.rowslice(feb_data_2018, 0, 19, 2)
```

```python
alternate_data_feb_2018.displayall()
```

Choose and/or re-order fields

```python
open_close = etl.cut(index_data,'Date', 'Open', 'Close')
```

```python
open_close.display()
```

```python
high_low = etl.cut(index_data, 0, 2, 3)
```

```python
high_low.display()
```

```python
ohlc = etl.cut(index_data, *range(0, 5))
```

```python
ohlc.display()
```

```python
nyse_data = etl.cutout(index_data, 'Volume')
```

```python
nyse_data.display()
```
