#### Using a context method
This filters out large upward changes in the FX rate. This is clearly a currency which swings quite wildly.

```python
fx_rates = [['date', 'close'],
            ['2019-08-01', 10],
            ['2019-08-02', 15],
            ['2019-08-03', 13],
            ['2019-08-04', 10],
            ['2019-08-05', 17],
            ['2019-08-06', 20],
            ['2019-08-07', 21],
            ['2019-08-08', 20]]
```

```python
def query(prv, cur, nxt):
     return ((prv is not None and (cur.close - prv.close) < 2)
     or (nxt is not None and (nxt.close - cur.close) < 2))
```

```python
fx = etl.selectusingcontext(fx_rates, query)
```
