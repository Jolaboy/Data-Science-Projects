```python
emp_details = etl.fromtext('./datasets/emp_details.txt', header = ['Details'] )
```

```python
etl.lookall(emp_details)
```

strip() function is called on each line, which by default will remove leading and trailing whitespace,
including the end-of-line character

#### Use the capture() method to extract field values
https://petl.readthedocs.io/en/stable/transform.html?highlight=capture#petl.transform.regex.capture


```python
emp_details_table = emp_details.capture('Details', '(.*),(.*),(.*)$',
                                        ['ID', 'Name', 'Department'])
```

```python
etl.lookall(emp_details_table)
```

```python
template = '''|->
              | {ID}
              | {Name}
              | {Department}'''
```

```python
epilogue = '|}'
```

```python
etl.totext(emp_details_table, 
           'emp_details_sample.txt', 
           template=template, 
           epilogue=epilogue)
```

```python
print(open('emp_details_sample.txt').read())
```

```python
prologue = '=========START=========\n'

template = '########\n'\
           '{ID} | {Name} | {Department}\n'

epilogue = '=========END========='
```

```python
etl.totext(emp_details_table, 
           'emp_details_sample.txt', 
           template=template,
           prologue=prologue,
           epilogue=epilogue)
```

```python
print(open('emp_details_sample.txt').read())
```
