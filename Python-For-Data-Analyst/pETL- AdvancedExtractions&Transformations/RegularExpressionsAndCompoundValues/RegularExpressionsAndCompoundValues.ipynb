#### Unpacking compound values
- None values can be included. Missing values are treated as None

```python
names_of_players = [['bioID', 'name'],
                    ['cheanca01', ['Calbert', 'Nathaniel', 'Cheaney']],
                    ['collija01', ['James', 'Edgar', 'Collins']],
                    ['evansea01', ['Earl', 'Joseph', 'Evans']],
                    ['rosegl01',  ['Glen', None, 'Rose']],
                    ['randy01',   ['Randy']]]
```

```python
names = etl.unpack(names_of_players, 'name', ['first name', 'middle name', 'last name'])
```

```python
names.display()
```

```python
dict_of_names = [['bioID', 'name'],
                 ['cheanca01', {'first_name':'Calbert', 'middle_name':'Nathaniel', 'last_name':'Cheaney'}],
                 
                 ['collija01', {'first_name':'James', 'middle_name':'Edgar', 'last_name':'Collins'}],
                 
                 ['evansea01', {'first_name':'Earl', 'middle_name':'Joseph', 'last_name':'Evans'}]]
```

```python
names = etl.unpackdict(dict_of_names, 'name')
```

```python
names.display()
```
